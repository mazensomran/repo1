```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = "/content/drive/MyDrive/augmented_leather_def/Defcut3.png"
```

```python
input_img = cv2.imread(image)  # Read images

img = cv2.resize(input_img,(100, 100))
blue_channel, green_channel, red_channel = cv2.split(img)

 # تحويل الصورة إلى مساحة لونية HSV
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h_channel, s_channel, v_channel = cv2.split(hsv_image)
 # تحويل الصورة إلى مساحة لونية LAB
lab_image = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l_channel, a_channel, b_channel = cv2.split(lab_image)

Chanels = {'blue_channel' :blue_channel, 'green_channel' : green_channel, 'red_channel': red_channel,
           'h_channel': h_channel, 's_channel': s_channel, 'v_channel': v_channel,
           'l_channel': l_channel, 'a_channel': a_channel, 'b_channel' : b_channel}

cv2_imshow(img)
print('original')
for k,v in Chanels.items():
  print(k)
  cv2_imshow(v)
```
